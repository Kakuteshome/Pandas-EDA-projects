```python
#Time to ask some questions
#1. what are the locations with the fastest roller coasters?(minimum of 10)
df['Location'].value_counts()
```

```python
 #By running this line of code, I'm grouping all of the rides as well as aggregating them by count and sum. I'm also filtering out the locations that have less than 10 rides. 
 ax3 =  df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean','count']).query('count>=10').sort_values('mean')
```

```python
#plot the above findings into a bar plot to share our insights
~ax3['count'].plot(kind='barh',figsize=(12,5),title='Average Coast Speed by Location')
plt.show()
```
